```python
import pandas as pd
```

```python
data = pd.read_csv("feature_name.txt", header = None, delimiter='\n')[7:]
```

```python
a = data[0].values
```

```python
a
```

```python
f = []
t = []

for i in a:
    if i[0] == 'f':
        f.append(i)
    else:
        t.append(i)
```

```python
temp = []
for j in f:
    q = j.split('-')
    if len(q) == 2:
        temp.append(q[1])
    else:
        temp.append('-'.join(q[1:]))   
```

```python
sorted(set(temp))

```
